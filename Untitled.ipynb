```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
dataframe = pd.read_csv('Resources/cities.csv')
```

```python
dataframe.head()
```

```python
humidity_lat = dataframe[["Humidity","Lat"]]
cloudiness_lat = dataframe[["Cloudiness","Lat"]]
maxtemp_lat = dataframe[["Max Temp","Lat"]]
windspeed_lat = dataframe[["Wind Speed","Lat"]]


# humidity_lat.head()
```

```python
humidity_lat.plot(x="Lat",y="Humidity",kind="scatter", figsize=(10,7))
plt.title("Vs Humidity")
plt.tight_layout()
plt.savefig('figures/vsHumidity.jpg' )
```

```python
cloudiness_lat.plot(x="Lat",y="Cloudiness", kind="scatter",figsize=(10,7))
plt.title("Vs Cloudiness")
plt.tight_layout()
plt.savefig("figures/vsCloudiness.jpg")
```

```python
maxtemp_lat.plot(x="Lat",y="Max Temp",kind="scatter",figsize=(10,7))
plt.title("Vs Max Temp")
plt.tight_layout()
plt.savefig("figures/vsMaxTemp.jpg")
```

```python
windspeed_lat.plot(x="Lat",y="Wind Speed", kind="scatter",figsize=(10,7))
plt.title("Vs Wind Speed")
plt.tight_layout()
plt.savefig("figures/vsWindSpeed.jpg")
```
